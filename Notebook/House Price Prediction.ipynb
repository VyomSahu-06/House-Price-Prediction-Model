```python
#Loading the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the Dataset
df=pd.read_csv("house_prices_practice.csv")
print(df)
```

```python
#Performing EDA
print("\n Columns:-\n",df.columns)
print("\n Shape:-\n",df.shape)
print("\nFirst 5 rows of the dataset:-\n",df.head())
print("\nLast 5 rows of the dataset:-\n",df.tail())
print("\nDescription:-\n",df.describe())
print("\n Correlation:-\n",df.corr(numeric_only=True))
print("\nMean:-\n",df.mean(numeric_only=True))
print("\nStandard Deviation:-\n",df.std(numeric_only=True))
```

```python
#Cleaning the Dataset
print(df.isna().sum())

#Removing unwanted Columns
df=df.drop(["Id"],axis=1)
print(df.isna().sum())

#Checking the datatype of all the columns
print(df.dtypes)

#As we can see the there is no missing values in the dataset so we know that the data is cleaned.
```

```python
#Visualizing the data
#Creating a histogram of prize distribution.
plt.figure(figsize=(6,10))
plt.hist(df["SalePrice"],bins=20,color="red")
plt.xlabel("Sale Price")
plt.ylabel("Number of Houses")
plt.title("House Price Distribution")
plt.show()
```

```python
#Plotting the scatter plot of area vs price
plt.figure(figsize=(10,6))
plt.scatter(df["GrLivArea"],df["SalePrice"],marker="o",linewidths=2,edgecolors="black",color="yellow")
plt.xlabel("Ground Living Area")
plt.ylabel("Sale Price")
plt.title("Living Area vs Sale Price")
plt.show()
```

```python
#Plotting the bar chart to find the average price by Overall Quality
qualavg=df.groupby("OverallQual")["SalePrice"].mean()
print(qualavg)
plt.figure(figsize=(10,6))
plt.barh(qualavg.index,qualavg.values,align="edge",color="green",edgecolor="black",linewidth=2)
plt.xlabel("Average Sale Price")
plt.ylabel("Overall Quality")
plt.title("Average Sale Price by Overall Quality")
plt.show()
```

```python
#Creating a box plot to calculate the garage cars by sale price
plt.figure(figsize=(10,3))
sns.boxplot(x="GarageCars",y="SalePrice",data=df,linewidth=2,color="purple",linecolor="black")
plt.xlabel("Garage Cars")
plt.ylabel("Sale Price")
plt.title("Garage Cars by Sale Price")
plt.show()
```

```python
#Creating a scatter plot for checking the sale price by years built
plt.figure(figsize=(10,5))
plt.scatter(df["YearBuilt"],df["SalePrice"],marker="o",linewidths=2,edgecolor="black",color="blue")
plt.xlabel("Years Built")
plt.ylabel("Sale Price")
plt.title("Sale Price by Year Built")
plt.show()
```

```python
#Creating a scatter plot for comparing the sale price by the area covered by basement
plt.figure(figsize=(10,6))
plt.scatter(df["TotalBsmtSF"],df["SalePrice"],linewidths=2,marker="o",edgecolor="black",color="orange")
plt.xlabel("Total Basement Area")
plt.ylabel("Sale Price")
plt.title("Basement Area vs Sale Price")
plt.show()
```

```python
#Plotting a pair plot for showing the most important variables together
plt.figure(figsize=(8,6))
sns.pairplot(df[[
    "SalePrice",
    "GrLivArea",
    "OverallQual",
    "GarageCars",
    "FullBath"
]])
plt.show()
```

```python
#Plottig the box plot for showing the outliers of the sale price
plt.figure(figsize=(10,6))
plt.boxplot(df["SalePrice"])
plt.ylabel("Sale Price")
plt.title("Sale Price Outliers")
plt.show()
```

```python
#Plotting a correlation heatmap of all the numeric values present in it
corr=df.corr(numeric_only=True)
print(corr)
plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True,cmap="plasma")
plt.show()
```

```python
#Feature Engineering
current_year=2026

#House Age
df["House Age"]=current_year-df["YearBuilt"]
#Total Size
df["Total Size"]=df["GrLivArea"]+df["TotalBsmtSF"]
#Area per Bedroom
df["AreaperBed"]=df["GrLivArea"]/df["BedroomAbvGr"]
#Bathrooms per Bedroom
df["BathperBed"]=df["FullBath"]/df["BedroomAbvGr"]

#Also dropping the year built columns as the house age and year built column do the exact same task
df.drop("YearBuilt",axis=1,inplace=True)
#After feature Engineering the dataset looks like this
print(df.head())
print(df.shape)

```

```python
#Encoding the categorical values to numerical variables
print(df.dtypes)
#As we can see all the columns are numerical so there is no need for Encoding.
```

```python
#Now converting the dataset into the csv file
df.to_csv("New House Price Prediction.csv")
```

```python
#Now splitting the dataset into train and test split
from sklearn.model_selection import train_test_split
x=df.drop(["SalePrice"],axis=1)
y=df["SalePrice"]

x_train,x_test,y_train,y_test=train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)

print(x_train)

#Now applying the linear regression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_predlr=lr.predict(x_test)

#Now applying Decision tree model
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=42)
dt.fit(x_train,y_train)
y_preddt=dt.predict(x_test)

#Now applying the Random Forset model
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
rf.fit(x_train,y_train)
y_predrf=rf.predict(x_test)

#Now Evaluating the model with the mae,mse,rmse and r2 score
from sklearn.metrics import (mean_absolute_error,mean_squared_error,r2_score)
models={
    "Linear Regression" : y_predlr,
    "Decision Tree" : y_preddt,
    "Random Forest" : y_predrf
}

for name,pred in models.items():
    print("="*40)
    print(name)
    print("="*40)
    mse=mean_squared_error(y_test,pred)
    mae=mean_absolute_error(y_test,pred)
    rmse=np.sqrt(mse)
    r2=r2_score(y_test,pred)
    print(f"Mean Absolute Error (MAE) : {mae:.6f}")
    print(f"Mean Squared Error (MSE) : {mse:.6f}")
    print(f"Root Mean Squared Error (RMSE) : {rmse:.6f}")
    print(f"R2 Score         : {r2:.4f}")
```

```python
#Now checking for the final prediction
new_house = pd.DataFrame({
    "OverallQual": [8],
    "GrLivArea": [2200],
    "GarageCars": [2],
    "TotalBsmtSF": [1100],
    "FullBath": [3],
    "BedroomAbvGr": [4],
    "LotArea": [9500],
    "House Age": [8],
    "Total Size": [3300],
    "AreaperBed": [550],
    "BathperBed": [0.75]
})
predicted=lr.predict(new_house)
print("="*40)
print("House Price Prediction")
print("="*40)
print(f"Predicted House Price: ₹{predicted[0]:,.0f}")
```
